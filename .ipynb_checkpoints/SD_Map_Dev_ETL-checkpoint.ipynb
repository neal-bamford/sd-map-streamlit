```python
## Load Borough Level Estimated Population Data

log.debug("Starting Borough Estimated Population Data ET not L")
## Staging 01 200101 to 201003 - Has City of London Borough data in it 
estimated_borough_population_2001_2022_df = mlib.csv_to_dataframe("./data/_original/population/estimated_borough_population_2001_2022.csv")

estimated_borough_population_2001_2022_df_melt = pd.melt(estimated_borough_population_2001_2022_df, id_vars    = ["LAD","LAD_NAME"]\
                                                                      , value_vars = ["2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022"]
                                                                      , var_name   = "YEAR"
                                                                      , value_name = "COUNT")

estimated_borough_population_2001_2022_df_melt = estimated_borough_population_2001_2022_df_melt.round({"COUNT": 0})
estimated_borough_population_2001_2022_df_melt["COUNT"] = estimated_borough_population_2001_2022_df_melt["COUNT"].astype(int)
staging_estimated_borough_population_2010_2022_df = estimated_borough_population_2001_2022_df_melt.set_index("YEAR")

staging_estimated_borough_population_2010_2022_df_rows = staging_estimated_borough_population_2010_2022_df.shape[0]
log.debug(f"staging_estimated_borough_population_2010_2022_df_rows:{staging_estimated_borough_population_2010_2022_df_rows}")
staging_estimated_borough_population_2010_2022_df.to_csv("./data/_original/population/staging_estimated_borough_population_2001_2022_df.csv")

log.debug("Finished Borough Estimated Population Data ET not L")

estimated_borough_population_2001_2022_df_melt
```
